```python
ratings = zomato_data["rate"].value_counts().to_frame().reset_index()
# Dropping rows with no ratings
ratings = ratings.drop([0, 45])
ratings["index"] = np.array([float(item.split("/")[0]) for item in ratings["index"]])
ratings = ratings.sort_values(by="index")
ratings
```

```python
plt.figure(figsize=(15, 10))
sns.barplot(x=ratings["index"], y=ratings["rate"], color="yellow")
plt.title("Ratings of Restaurants", fontsize=25)
plt.xlabel("Rating", fontsize=15)
plt.ylabel("Number of restaurants", fontsize=15)
plt.gca().yaxis.grid(True)
```

```python
## Udupi in name
zomato_data.loc[zomato_data["name"].str.contains("Udupi")]
```

```python
highly_rated_by_category = zomato_data[zomato_data.rate.isin(["4.9/5", "4.8/5", "4.7/5"])]["rest_type"].value_counts().to_frame().reset_index()
highly_rated_by_category
```

```python
highly_rated_by_category["pct"] = (100. * highly_rated_by_category["rest_type"] / highly_rated_by_category["rest_type"].sum()).round(2)
```

```python
plt.figure(figsize=(10, 10))
sns.barplot(x=highly_rated_by_category["pct"], y=highly_rated_by_category["index"], palette="inferno")
plt.gca().xaxis.grid(True)
plt.title("Percentage of highly rated restaurants by type in Bangalore", fontsize=25)
plt.xlabel("Percentage of Restaurants", fontsize=15)
plt.ylabel("Name", fontsize=15)
```

Although there are lot many Quick Bites restaurants, Bangalore prefers Casual Dining as per the percentage of ratings. And of course, they love their Pubs, Bars and Breweries!

```python
cost_calculation_data = zomato_data[["name", "rest_type", "location", "approx_cost(for two people)"]]
cost_calculation_data
```

```python
costs = cost_calculation_data["approx_cost(for two people)"].value_counts().to_frame().reset_index().rename(columns={"index": "Cost", "approx_cost(for two people)": "Count"})
costs
```

```python
costs["pct"] = (100. * costs["Count"] / costs["Count"].sum()).round(2)
plt.figure(figsize=(10, 10))
sns.barplot(x=costs["pct"][:20], y=costs["Cost"][:20], palette="inferno")
plt.gca().xaxis.grid(True)
plt.title("Percentage of restaurants by price in Bangalore", fontsize=25)
plt.xlabel("Percentage of Restaurants", fontsize=15)
plt.ylabel("Name", fontsize=15)
```

```python
cuisines_df = zomato_data["cuisines"].dropna().to_list()
cuisines_df
```

```python
cuisines_ind = [item.split(", ") for item in cuisines_df]
cuisines = [item for sub_list in cuisines_ind for item in sub_list]
cuisines = list(set(cuisines))
cuisines
```

```python
cuisines_count_dict = {}
for cuisine in cuisines:
    for item in cuisines_df:
        if cuisine in item:
            if cuisine in cuisines_count_dict:
                cuisines_count_dict[cuisine] += 1
            else:
                cuisines_count_dict[cuisine] = 1
cuisines_count_dict
```

```python
cuisines_count_df = pd.DataFrame({"Cuisines": list(cuisines_count_dict.keys()), "Count": list(cuisines_count_dict.values())}).sort_values(by="Count", ascending=False).reset_index()
cuisines_count_df
```

```python
cuisines_count_df["pct"] = (100. * cuisines_count_df["Count"] / cuisines_count_df["Count"].sum()).round(2)
cuisines_count_df
```

```python
plt.figure(figsize=(10, 10))
sns.barplot(x=cuisines_count_df["pct"][:30], y=cuisines_count_df["Cuisines"][:30], palette="inferno")
plt.gca().xaxis.grid(True)
plt.title("Percentage of restaurants that serve cuisines in Bangalore", fontsize=25)
plt.xlabel("Percentage of Restaurants", fontsize=15)
plt.ylabel("Cuisine", fontsize=15)
```

More restaurants mention that they serve North Indian food than South Indian food. More Pizza and Burger places than those serving Seafood. As expected, Biryani comes pretty high on the list. Chinese makes to the top as well.

```python
plt.figure(figsize=(10, 10))
sns.barplot(x=cuisines_count_df["pct"][-50:], y=cuisines_count_df["Cuisines"][-50:], palette="inferno")
plt.gca().xaxis.grid(True)
plt.title("Cuisines served by the least number of restaurants in Bangalore", fontsize=25)
plt.xlabel("Percentage of Restaurants", fontsize=15)
plt.ylabel("Cuisine", fontsize=15)
```

Not surprisingly, all the exotic cuisines make the bottom of the list.
